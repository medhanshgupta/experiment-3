```python
tpl7=("apple","banana","orange")
print("before: ",id(tpl7))
tpl8=("kiwi",)
tpl7+=tpl8
print(tpl7)
print("after: ",id(tpl7))
```

```python
tpl7=["apple","banana","orange"] #list used the same memory location
print("before: ",id(tpl7))
tpl8=["kiwi",]
tpl7+=tpl8
print(tpl7)
print("after: ",id(tpl7))
```

```python
# marks of 5 students using tuple
marks=(90,82,66,76,33)
print("maximum marks: ",max(marks))
print("minimum marks: ",min(marks))
print("total number of students: ",len(marks))
print("sum of marks: ",sum(marks))
print("average marks: ",sum(marks)/len(marks))

print("sorted marks: ",sorted(marks))
```

```python
# student result
result=("maths",82,"A")
subject, marks, grade = result
print("subject: ",subject)
print("marks: ",marks)
print("grade: ",grade)
if marks>=75:
  print("distinction")
```

```python
attendance = ("p","a","p","p","a","p")
print("no. of days present: ",attendance.count("p"))
print("no. of days absent: ",attendance.count("a"))
if attendance.count("a")>=1:
  print("was absent at least once")
```
